```python
# Download the data using Pandaâ€™s DataReader() function

from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create one dataset per company and collect the data for the year 2021 (from January 1, 2021 to December 31, 2021)

start = datetime.datetime(2021, 1, 1)
end = datetime.datetime(2021, 12, 31)

WMT = data.DataReader("WMT", "yahoo", start, end)
TGT = data.DataReader("TGT","yahoo", start, end)
```

```python
# Create the columns %-chg and Log Returns for both companies.

WMT['Log Returns'] = np.log(WMT['Close']/WMT['Close'].shift())
TGT['Log Returns'] = np.log(TGT['Close']/TGT['Close'].shift())
```

```python
# Create a combined plot of the volumes for both companies. 

fig, ax = plt.subplots(2)

WMT['Volume'].loc['2021-01-01':].plot(ax=ax[0], title = "Walmart Volume 2021", figsize=(12,6))
TGT['Volume'].loc['2021-01-01':].plot(ax=ax[1], title = "Target Volume 2021", figsize=(12,6))

fig.tight_layout()

```

```python
# On the line right below the plot comment on the volume of both companies 
# (which volume is highest, what months have the highest volumes in both companies, etc.)

# We can see that walmart has higher volume than target in general.
# The busiest months of 2021 seem to be in March, June, and December.
# They both appear to follow very similar cycles.
```

```python
# Create a combined plot of the log returns on both companies and on the line right below comment and compare both returns.

fig, ax = plt.subplots(1,2)
WMT['Log Returns'].hist(ax=ax[0], bins=50, alpha=0.6, color='b', figsize=(12,4))
ax[0].set_ylabel("Freq of log return")
ax[0].set_title("WMT Returns ")


TGT['Log Returns'].hist(ax=ax[1], bins=50, alpha=0.6, color='b', figsize=(12,4))
ax[1].set_ylabel("Freq of log return")
ax[1].set_title("TGT Returns")

fig.tight_layout()
```

```python
# We can see that walmart returns is slightly more skewed right than target. Target is a tiny bit more normal between the two.
```
